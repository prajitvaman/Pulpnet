3. Print a downward Half-Pyramid Pattern

0 0 0 0 0   
0 0 0 0 <br>
0 0 0 <br>
0 0 <br>
0

```python
# we can see the iterator is decreasing with each execution, so I have employed a while loop to print and decrease the iterator by 1 in each step.
i = 5
while i>0:
  print('0'*i)
  i = i-1
```

 4. Given the following assignment of the vegetables list, add "tomato" to the end of the list and sort them in alphabetical order.\
vegetables = ["eggplant", "broccoli", "carrot","cauliflower", "zucchini"]

```python
# key idea: .append() method of the list class is used to add an item at the end of the list
vegetables = ["eggplant", "broccoli", "carrot","cauliflower", "zucchini"]
vegetables.append('tomato')
vegetables.sort()
vegetables

# alternate: "tomato" is a list of length = 1, concatenated to the list of vegetables.
vegetables = ["eggplant", "broccoli", "carrot","cauliflower", "zucchini"]
vegetables = vegetables + ["tomato"]
vegetables.sort()
vegetables
```

5. Write a function definition named is_odd that takes in a number and returns True or False if that number is odd.

```python
def is_odd(num):
  if num%2 == 1:
    return True
  else:
    return False

is_odd(5)
```

6. Write a function definition named mode that takes in sequence of numbers and returns the most commonly occuring value

```python
def mode(seq):
  return max(set(seq), key = seq.count)

seq = [1,2,2,2,3,4,4,5,6]
mode(seq)
```

Exercise 1: Create a 4X2 integer array and Prints its attributes \\
**Note:** The element must be a type of unsigned int16. \\
And print the following Attributes: â€“



1.   The shape of an array.
2.   Array dimensions.
3.   The Length of each element of the array in bytes.







```python
import numpy

arr = numpy.array([[1,2],[3,4],[5,6],[7,8]], dtype = numpy.uint16)
print(f'size of array: {arr.size}')
print(f'shape of array: {arr.shape}')
print(f'dimensions of array: {arr.ndim}')
print(f'length of each element in bytes: {arr.itemsize}')
```

Exercise 2: Following is the provided numPy array. Return array of items by taking the third column from all rows \\
sampleArray = numpy.array ( [ [ 11 ,22, 33 ], [ 44, 55, 66 ], [ 77, 88, 99 ] ] )

```python
sampleArray = numpy.array([[11,22,33],[44,55,66],[77,88,99]])
returnArray = numpy.array([])
upperLimit = sampleArray.shape[0]
for i in range(1, upperLimit+1):
  returnArray = numpy.append(returnArray, sampleArray[i-1][2])
returnArray
```

Exercise 2: Replace all column values which contain ?, n.a, or NaN with suitable values and print the updated dataset.:

```python
columns = list(df.columns)
for col in columns:
  df[col] = df[col].replace('?', numpy.nan)
  df[col] = df[col].replace('n.a', numpy.nan)

df
```
