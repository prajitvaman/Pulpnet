```python
def mode(lst):
  lst.sort()
  count,dum,mode=0,1,lst[0]
  for i in range(0,len(lst)-1):
    if lst[i]==lst[i+1]:
      dum=dum+1
    else:
      if count<dum:
        count=dum
        mode=lst[i]
      dum=1
  return mode

lst=[2,3,4,8,1,2,6,0]
print(mode(lst))
```

```python
col=df.columns
for i in col:
  for j in range(0,len(df)):
    if df[i][j]=='?' or df[i][j]=='n.a' or df[i][j]=='NaN':
      df[i][j]=0
df.head()
```

```python
car=df.company.unique()
arr=df.company
lst=np.zeros(len(car))
for i in range(0,len(car)):
  for j in arr:
    if car[i]==j:
      lst[i]=lst[i]+1
for i in range(0,len(car)):
  print(car[i],lst[i])
```
